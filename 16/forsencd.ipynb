```python
import sys, re
lines = [re.split('[\\s=;,]+', x) for x in raw.split("\n")]
```

```python
# GRAPH
G = {x[1]: set(x[10:]) for x in lines}
G
```

```python
# FLOW RATES
F = {x[1]: int(x[5]) for x in lines if int(x[5]) != 0}
F
```

```python
# EVIL BIT HACKING
# The key here is that for all nodes that have flow rates,
# I is a bitflag that says is it enabled or not
I = {x: 1<<i for i, x in enumerate(F)}
I
```

```python
# Floyd-Warshall
T = {x: {y: 1 if y in G[x]
              else float('+inf') 
            for y in G} 
         for x in G}

for k in T:
    for i in T:
        for j in T:
            T[i][j] = min(T[i][j], T[i][k]+T[k][j])
            
T
```

```python
def visit(v,       # the current node
          budget,  # the remaining time
          state,   # what things are currently enabled?
          flow,    # the current score
          answer   # the key thing is that hashmaps are mutated by reference in python
                   # 
         ):
    
    # this method works by SIGNIFICANTLY cutting down the search space, and then
    # visiting every possible order
    
    # answer
    #   the state is the bitflag
    #   the answer in the end is that best possible score we ever had in this state
    #   the additional argument to get (0) is a default value. this is equiv to
    #   `getWithDefault` in haskell
    answer[state] = max(answer.get(state, 0), flow)
    
    # we dont have to check neighbors after FW, we can visit any
    # node in F i. e. the nodes that have flow
    for u in F:
        # subract the time it takes to go to u
        newbudget = budget - T[v][u] - 1
        newflow   = flow + newbudget * F[u]
        
        # key insight: in a correct solution we dont visit anything twice!
        # there is an edge between everything, so we literally only go somewhere once
        
        # if bitfag&state we have visited before and dont go
        # if we dont have time to go here, we obviously dont go
        if I[u] & state or newbudget <= 0: continue
            
        # visit recursively
        visit(u, newbudget, state | I[u], newflow, answer)
        
    return answer    
```
